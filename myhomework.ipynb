```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns   
```

```python
data= pd.read_csv('movies.csv')
data
```

```python
data.info()
```

```python
data.describe()
```

```python
data.head(8)
```

```python
data.tail(9)
```

```python
#droping the unnamed column 
data=data.drop(columns=['Unnamed: 0'])
data
```

```python
#changing the release date to datetime format
data['release_date']=pd.to_datetime(data['release_date'])
data.info()
```

```python
#showing the first 10 rows of both the title and release date columns
data1=data[['title','release_date']].head(10)
data1
```

```python
#alternative way of showing the first 10 rows of both the title and release date columns
data2=data.loc[:,['title','release_date']].head(10)
data2
```

```python
#giving conditions and adding a new column to it
data.loc[data['vote_average']<5,'ratings']='fair'
data.loc[data['vote_average']>5,'ratings']='good movie'
data
```

```python
#using lambda
data['is_popular'] = data['popularity'].apply(lambda x: 'great movie' if x > 20 else 'sucks')
data
```

```python
#correlation between vote count and popularity
data[['vote_count','popularity']].corr()
```

```python
#plotting a line plot for popularity against vote count using matplotlib
plt.figure(figsize=(9,6))
plt.plot(data['popularity'],data['vote_count'],color='purple')
plt.title('popularity vs vote count')
plt.xlabel('popularity')
plt.ylabel('vote count')
plt.show()
```

```python
data3=data[['popularity','vote_average']].head(10)
data3
```

```python
#line
plt.figure(figsize=(9,6))
plt.scatter(data3['popularity'],data3['vote_average'],color='purple')
plt.title('popularity vs vote average')
plt.xlabel('popularity')
plt.ylabel('vote count')
plt.show()
```

```python
data4=data[['popularity','ratings']].head(20)
data4   
```

```python
#plotting a box plot for popularity against ratings using seaborn
plt.figure(figsize=(10,8))
sns.boxplot(x='ratings',y='popularity',data=data4)  
plt.title('popularity vs ratings',fontsize=17,color='green',fontweight='bold',)
plt.xlabel('ratings')       
plt.ylabel('popularity')
plt.show()

```
